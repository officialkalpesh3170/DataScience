```python
import requests
from bs4 import BeautifulSoup

import pandas as pd


from tqdm import tqdm
```

```python
data=[]
for pages in tqdm(range(1,51)):
  link=f'https://books.toscrape.com/catalogue/page-{pages}.html'
  req=requests.get(link)
  soup=BeautifulSoup(req.text,'html.parser')
  for sp in soup.find_all('li' ,class_="col-xs-6 col-sm-4 col-md-3 col-lg-3"):
     img_link=sp.find('img').get('src')
     link1=sp.find_all('a')[-1].get('href')
     title=sp.find_all('a')[-1].get('title')
     rating=sp.find('p').get('class')[-1]
     price=sp.find('p',class_='price_color').text
     price=float(price[2:])
     stock=sp.find('p',class_='instock availability').text.strip()
     data.append([title,link1,price,stock,rating,img_link])




```

```python
df=pd.DataFrame(data,columns=['Title','Link','Price','Stock','Rating','img_link'])
```

```python
df.head()
```

```python
rating_mapping = {
    'One': 1,
    'Two': 2,
    'Three': 3,
    'Four': 4,
    'Five': 5
}

# Apply the mapping to the 'Rating' column
df['Rating'] = df['Rating'].map(rating_mapping)

# Now you can convert the 'Rating' column to integers
df['Rating'] = df['Rating'].astype(int)
```

```python
len(df)
```

```python
df.to_csv('books.csv',index=False)
```

```python
base_url = 'https://books.toscrape.com/catalogue/'  # Define the base URL
data=[]
for i in tqdm(df1['Link']):
  full_url = base_url + i  # Construct the full URL
  res = requests.get(full_url)
  soup = BeautifulSoup(res.text, 'html.parser')

  type=soup.find('ul',class_='breadcrumb').find_all('a')[2].text
  upc=soup.find('table',class_='table table-striped').find_all('td')[0].text
  price_x=soup.find('table',class_='table table-striped').find_all('td')[2].text[2:]
  price_i=soup.find('table',class_='table table-striped').find_all('td')[3].text[2:]
  tax=soup.find('table',class_='table table-striped').find_all('td')[4].text[2:]
  avail=soup.find('table',class_='table table-striped').find_all('td')[5].text
  qn=soup.find('table',class_='table table-striped').find_all('td')[6].text

  data.append([type,upc,price_x,price_i,tax,avail,qn])
```

```python
len(data)
```

```python
df=pd.DataFrame(data,columns=['type','upc','price_x','price_i','tax','avail','qn'])
```
