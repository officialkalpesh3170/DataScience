````python
def anagram(a,b):
  a=a.lower()
  b=b.lower()

  sort_a= ''.join(sorted(a))
  sort_b=''.join(sorted(b))

  if(sort_a==sort_b):
    return True

  else:
    return False



a='CAT'
b='Tac'


print(anagram(a,b))


````
